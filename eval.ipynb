```python
import anthropic

client = anthropic.Anthropic()
```

```python
def get_prompt(query:str, answer:str, generated_answer:str) -> str:
    prompt = """Given following MathVista query, answer and generated answer, output "yes" or "no" regarding whether the generated answer is correct.
    Query: <query> {query} </query>
    Answer: <answer> {answer} </answer>
    Generated Answer: <generated_answer> {generated_answer} </generated_answer>

    Output only "yes" or "no" regarding correctness of the generated answer.
    Output:""".format(query=query, answer=answer, generated_answer=generated_answer)
    return prompt
```

```python
import pandas as pd

paligemma2_base_results = pd.read_csv("results.csv")
print(paligemma2_base_results.iloc[0])
```

```python
validity_list = []
for i in range(len(paligemma2_base_results)):
    print(f"step {i}")
    message = client.messages.create(
        model="claude-3-5-sonnet-20241022",
        max_tokens=1000,
        temperature=0,
        messages=[
            {
                "role": "user",
                "content": [
                    {
                        "type": "text",
                        "text": get_prompt(
                            query=paligemma2_base_results.iloc[i]["prompt"], 
                            answer=paligemma2_base_results.iloc[i]["true_answer"], 
                            generated_answer=paligemma2_base_results.iloc[i]["result"])
                    }
                ]
            }
        ]
    )
    validity_list.append(message.content[0].text)
```

```python
paligemma2_base_results["validity"] = validity_list
print(f"Percent right: {round(len(paligemma2_base_results[paligemma2_base_results["validity"] == "yes"]) / len(paligemma2_base_results) * 100, 2)}%")
```
